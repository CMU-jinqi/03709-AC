```python
import matplotlib.pyplot as plt

# dataset I need to predict
methods = ["Full", "AF2-only", "Filters-only"]
success_rate = [90, 65, 35]

# plot
plt.figure(figsize=(6, 4), facecolor="white")  #
bars = plt.bar(methods, success_rate, color=["#4B8BBE", "#FFD43B", "#646464"], edgecolor="none")

# label
plt.title("Expected Success Rate by Design Method")
plt.xlabel("Design Method", fontsize=11)
plt.ylabel("Success Rate (%)", fontsize=11)
plt.ylim(0, 100)

# edit
for spine in ["top", "right"]:
    plt.gca().spines[spine].set_visible(False)
plt.grid(False)
plt.tight_layout()
plt.show()
```
